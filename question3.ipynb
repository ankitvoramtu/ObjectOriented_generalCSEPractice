# Prims Algorithm

### Steps

1. Choose random vertex 
2. Make table of weight, vertex, vertex
    * remove repeats
2. choose cheapest edge from chosen vertices
    * requires function: fun(chosenVertices): return cheapest edge
    * if if all vertices are in chosen vertices, break 
    * if cheapest edge is connected to a vertex in chosen vertices, then choose next cheapest
    * add vertex to chosen vertices
3. 

```python
import random
from IPython.display import Image, display

chosenVertex = []
chosenEdges = []

s32 = {'A':[('E',5),('H',6),('F',1),('B',8),('B',8)],'B':[('F',6),('C',4)],'C':[('F',2),('G',7)],'G':[('F',9)],'F':[('H',5)],'H':[('E',3)]}
chosenVertex.append(random.choice(s32.keys()))
#chosenVertex.append('A')
print chosenVertex
```

## Make table of weight, vertexa, vertexb 
### remove repeats and order by weights

```python
weightTable = []
for key in s32:
    for val in s32[key]:
        v = sorted([key, val[0]])
        if [val[1],v[0],v[1]] not in weightTable:
            weightTable.append([val[1],v[0],v[1]]) 
weightTable =  sorted(weightTable)
print weightTable
completeVertexes = sorted(list(set([key for key in s32 for val in s32[key] ] + 
                            [val[0] for key in s32 for val in s32[key]])))
```

## choose cheapest edge from chosen vertices
### new vertex cannot be inside chosen vertices

```python
while(chosenVertex != completeVertexes):   
    for val in weightTable:
        if val[1] in chosenVertex or val[2] in chosenVertex:
            if val[1] not in chosenVertex or val[2] not in chosenVertex:
                if val not in chosenEdges:
                    chosenEdges.append(val)
                chosenVertex.append(val[2] if val[2] not in chosenVertex else val[1])
                chosenVertex = sorted(chosenVertex)
                break
        
print 'chosenEdges = {} \nchosenVertex = {}'.format(chosenEdges, chosenVertex)
```
